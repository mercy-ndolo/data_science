```python
import matplotlib.pyplot as plt

x = ["bus", "car", "bicycle", "walking", "train", "motorcycle"]
y1 = [22, 11, 9, 24, 17, 17]
y2 = [13, 46, 1, 4, 34, 2]

# Set the width of the bars
bar_width = 0.35

# Calculate the x positions for the bars
x1 = range(len(x))
x2 = [x_val + bar_width for x_val in x1]

# Create the bar plots
plt.bar(x1, y1, label="1960", width=bar_width)
plt.bar(x2, y2, label="2010", width=bar_width)

# Set x-axis labels and ticks
plt.xlabel("Means of Transport")
plt.xticks([x_val + bar_width / 2 for x_val in x1], x)

# Set y-axis label
plt.ylabel("Kilometres")

# Set the plot title and legend
plt.title("Commuter Travel in Tokyo")
plt.legend()

# Show the plot
plt.show()
```
